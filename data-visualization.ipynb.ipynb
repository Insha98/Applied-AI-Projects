```python
import numpy as np
x=np.array([2020,2021,2022,2023])
y1=np.array([45,34,46,66])
y2=np.array([47,36,48,69])
y3=np.array([29,50,60,98])
y4=np.array([99,40,78,99])
plt.plot(x,y1,marker="*",
       markerfacecolor="black",
       ms=30,color="red",
        linestyle="dotted"
        )
plt.plot(x,y2,marker="*",
       markerfacecolor="black",
       ms=30,color="blue",
        linestyle="dotted"
        )
plt.plot(x,y3,marker="*",
       markerfacecolor="black",
       ms=30,color="green",
        linestyle="dotted"
        )
plt.plot(x,y4,marker="*",
       markerfacecolor="black",
       ms=30,color="orange",
        linestyle="dotted"
        )
```

```python

x=np.array([2020,2021,2022,2023])
y1=np.array([45,34,46,66])
y2=np.array([47,36,48,69])
y3=np.array([29,50,60,98])
y4=np.array([99,40,78,99])
edit=dict(marker="*",
       markerfacecolor="black",
       ms=20,
        linestyle="dotted"
        )
plt.plot(x,y1,**edit,
         color="blue")
plt.plot(x,y2,**edit,
         color="red")
plt.plot(x,y3,**edit,
         color="green")
plt.plot(x,y4,**edit,color="orange")
plt.grid(color="black",linewidth=1)
plt.title("students")
plt.xlabel("year")
plt.ylabel("no of students")
plt.xticks([2020,2021,2022,2023])#converting to int
plt.tick_params(axis="both",colors="violet")

```

```python
#bar graph =shows particular values
fruits=["apple","mango","pear","grapes"]
values=[10,20,30,45]
plt.bar(fruits,values,color="green")
plt.title("bar graph")
plt.xlabel("fruits")
plt.ylabel("values")
```

```python
#pie chart used for percentage
kashmir=["srinagar","gulmarg","pahalgam","sonamarg"]
colors=["green","blue","red","orange"]
values=[10,20,30,40]
plt.pie(values,labels=kashmir,colors=colors,autopct="%1.1f%%",explode=([0.1,0,0.1,0]))

```

```python
#subplot;the figure and axis
x=np.array([1,2,3,4,5,6])
figure,axes=plt.subplots(2,2)
axes[0,0].plot(x,x**2)
axes[0,1].plot(x,x**5)
axes[1,0].plot(x,x**3)
axes[1,1].plot(x,x**4)
```
