```python
from abc import ABC,abstractmethod
class Vehicle(ABC):
    @abstractmethod
    def start(self):
        print("vbn")
        @abstractmethod
        def stop(self):
             print("gn")
class MotorCycle(Vehicle):
    def start(self):
        print("motorcycle has started")
    def stop(self):
        print("stopped")
class Car(Vehicle):
    def start(self):
        print("car has started")
    def stop(self):
        print("stopped")
mobj = MotorCycle()
mobj.start()
mobj.stop()
c=Car()
c.start()
c.stop()
v=[MotorCycle(),Car()]
for i in v:
    i.start()
    i.stop()
            
```

```python
class Shape:
    def __init__(self,color,filled):
       self.color = color
       self.filled = filled
class Square(Shape):
    def __init__(self,color,filled,radius):
      super().__init__(color,filled)
      self.radius= radius
class Traingle(Shape):
    def __init__(self,color,filled,side):
       super().__init__(color,filled)
       self.side=side
s=Traingle("red",True,20)
print(s.color)
print(s.filled)
print(s.side)
```
