```python
#model training
import pandas as pd #loading dataset,organize data ,reading data
import numpy as np
import matplotlib.pyplot as plt #data visualization
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
```

```python
data=pd.read_csv("1771394594414_student_data.csv")
data
```

```python
data.info()
```

```python
plt.scatter(data["Study_Hours"],data["Marks"])
plt.xlabel("Study_Hours")
plt.ylabel("Marks")
plt.show()
```

```python
x=data[["Study_Hours"]]#using double paranthesis only bcz to act as a dataframe
y=data["Marks"]
```

```python
x_train,x_test,y_train,y_test=train_test_split(
    x,y,
    test_size=0.2,
    random_state=40
)
```

```python
model = LinearRegression() 
model.fit(x_train,y_train)
```

```python
#predictions
y_pred=model.predict(x_test)
```

```python
y_pred
```

```python
plt.scatter(x_test,y_pred)

```

```python
#model cleaning
r2_score(y_test,y_pred)
```
